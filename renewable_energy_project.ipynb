```python
wdata = df[(df['Country'].isin(['France','Germany'])) & (df['Energy Type'].isin(['Solar','Wind']))]
```

```python
wdata = wdata.drop(['Renewable Energy Jobs','Government Policies','R&D Expenditure','Renewable Energy Targets','Annual Rainfall','Public Awareness','Political Stability','Educational Level','Hydro Potential','Geothermal Potential','Biomass Availability','Energy Storage Capacity','Grid Integration Capability','Energy Subsidies','Energy Efficiency Programs','Urbanization Rate','Industrialization Rate','Energy Market Liberalization','Renewable Energy Patents','Technology Transfer Agreements','Renewable Energy Education Programs','Local Manufacturing Capacity','Import Tariffs on Energy Equipment','Export Incentives for Energy Equipment','Natural Disasters','Corruption Perception Index','Regulatory Quality','Rule of Law','Control of Corruption','Economic Freedom Index','Ease of Doing Business','Innovation Index','Number of Research Institutions','Number of Renewable Energy Conferences','Number of Renewable Energy Publications','Energy Sector Workforce','Public-Private Partnerships in Energy','Regional Renewable Energy Cooperation'], axis=1) 
```

```python
wdata.reset_index(drop =True)
```

```python
wdata = wdata.rename(columns={'Country':'country',
                              'Year' : 'year',
                              'Energy Type': 'energy_type',
                              'Production (GWh)' : 'production_gwh',
                              'Installed Capacity (MW)': 'installed_capacity_mw',
                              'Investments (USD)' : 'investments_usd',
                              'Population' : 'population',
                              'GDP' : 'gdp',
                              'Energy Consumption' : 'energy_consumption',
                              'Energy Exports' : 'energy_exports',
                              'Energy Imports' : 'energy_imports',
                              'CO2 Emissions' : 'co2_emissions',
                              'Average Annual Temperature' : 'avarage_annual_tremperature',
                              'Solar Irradiance' : 'solar_irradiance',
                              'Wind Speed' : 'wind_speed',
                              'Electricity Prices' : 'electricity_prices',
                              'International Aid for Renewables' : 'international_aid_for_renewables',
                              'Proportion of Energy from Renewables' : 'proportion_of_energy_from_renewables'
                             })
```

```python
wdata.head()
```

```python
sns.pairplot(wdata)
```

```python
p = sns.catplot(x ='year', y='production_gwh', data=wdata, hue='country' ,kind = 'bar', col= 'energy_type', palette = 'rocket', errorbar=None)
plt.rcParams['axes.formatter.limits'] = (-99, 99)
p.set_xticklabels(rotation=45)
plt.show()
```

```python
d = sns.catplot(x = 'year', y = 'energy_consumption', data = wdata, hue = 'country', kind = 'bar', col= 'energy_type', errorbar = None, palette= 'bright')
plt.rcParams['axes.formatter.limits'] = (-99, 99)
d.set_xticklabels(rotation=45)
```

```python
plt.figure(figsize=(10, 8))
sns.lineplot(x ='year', y ='co2_emissions', data = wdata, hue = 'country', errorbar = None, palette= 'bright')
plt.title('CO2 emission per year by country')
```

```python
p = sns.catplot(x ='year', y='investments_usd', data=wdata, hue='country' ,kind = 'bar', col= 'energy_type', palette = 'Paired', errorbar=None)
plt.rcParams['axes.formatter.limits'] = (-99, 99)
p.set_xticklabels(rotation=45)
plt.show()
```
